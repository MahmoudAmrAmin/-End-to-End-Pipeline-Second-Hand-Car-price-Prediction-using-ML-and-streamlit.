# handel missing values

- ### iterate through each column and calculate `number` of missing values and percent of each column 

```python
""" 
    Mahmoud implementation 

"""
# for col in df.columns : 
#     print(f'number of missing values is {df[col].isna().sum()} and percent of missing values in this column is {df[col].isna().sum() / len(df)}') 


"""
    implementation of mehrdad
"""
for col in df.columns : 
    number_of_null = df.loc[: , col].isna().sum() 
    perc_null = (number_of_null / df.shape[0] ) * 100 

    print(f'in column {col} - {number_of_null} - {perc_null}%')
```

```python
duplicate_rows = df.duplicated() 
if duplicate_rows.any() : 
    print('There is Duplicated') 
else : 
    print('No Duplication')    
```
